```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SpatialDropout1D, LSTM, Bidirectional, Dense
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('/content/Preprocessed_Data.csv')
df = df.drop('label',axis=1)
df
```

```python
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['Sentiment'])
```

```python
df
```

```python
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['sentence'], df['label'], test_size=0.2, random_state=42, stratify=df['label']
)
```

```python
max_words = 10000
max_len = 200
```

```python
tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(train_texts)
```

```python
train_sequences = tokenizer.texts_to_sequences(train_texts)
test_sequences = tokenizer.texts_to_sequences(test_texts)
```

```python
train_padded = pad_sequences(train_sequences, maxlen=max_len, padding='post')
test_padded = pad_sequences(test_sequences, maxlen=max_len, padding='post')
```

```python
train_labels = to_categorical(train_labels, num_classes=3)
test_labels = to_categorical(test_labels, num_classes=3)
```

```python
model = Sequential()
model.add(Embedding(max_words+1, 100, input_length=max_len))
model.add(SpatialDropout1D(0.2))
model.add(Bidirectional(LSTM(128, return_sequences=False, recurrent_dropout=0.2)))
model.add(Dense(128, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(
    train_padded, train_labels,
    validation_data=(test_padded, test_labels),
    epochs=15,
    batch_size=32,
    verbose=1
)
```

```python
eval_results = model.evaluate(test_padded, test_labels)
print(f"Test Loss: {eval_results[0]}, Test Accuracy: {eval_results[1]}")
```

```python
model.save('bilstm_sentiment_model.h5')
```
